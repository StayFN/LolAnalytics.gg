```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
dict_matches = {}
list_of_players = []

with open("../data/dict.txt") as f:
    for line in f:
        dict_matches[line.split(":", 1)[0]] = line.split(":", 1)[1].replace(" ", "").split(",")
    f.close()


for match_id in dict_matches.keys():
       list_of_players = list_of_players + dict_matches[match_id][:-1]

list_of_players_unique = list(set(list_of_players))
list_of_players_unique = list_of_players_unique[1:] #Removing "" entry as element

print(f'Length of list of players: {len(list_of_players)}')
print(f'Length of list of players distinct: {len(list_of_players_unique)}')
```

```python
print(list_of_players_unique)
```

```python
player_dict = {}
for player_id in list_of_players_unique[:50]:
    player_dict[player_id] = []

#Takes about 5 mins
for player_id in player_dict.keys():
    for match_id in dict_matches.keys():
        if player_id in dict_matches[match_id]:

            match_of_player = dict_matches[match_id][:-1]
            match_of_player.remove(player_id)

            player_dict[player_id] = list(set(player_dict[player_id] + match_of_player))
```

```python
# d = {0: {1, 2, 0}, 1: {2}}
G = nx.Graph(player_dict)

plt.figure(1, figsize=(20,10))
nx.draw_networkx(G, with_labels=False, node_size=10)
```
